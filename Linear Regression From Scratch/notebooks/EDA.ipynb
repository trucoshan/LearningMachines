## Visualizations and EDA

```python
for i in ['area','bedrooms','bathrooms','stories','parking']:
    plt.figure(figsize=(8,6))
    sns.lineplot(data=df,x=i,y='price')
    plt.tight_layout()
    plt.savefig(f"{i}.png",dpi=300,format='png')
    plt.show()
```

```python
for i in ['mainroad','guestroom','hotwaterheating','airconditioning','prefarea','furnishingstatus']:
    
    plt.figure(figsize=(8,6))
    sns.barplot(data=df,x=i,y='price',estimator=np.mean)
    plt.tight_layout()
    plt.savefig(f"{i}.png",dpi=300,format='png')
    plt.show()
```

## Final Observations

- The price increases with increase in area, number of bathrooms, bedrooms, parking spaces, and stories.
- The price is also more for properties with certain amenities.

## Final thought

**This dataset is perfect for testing out my linear regression model**

```python
df = df[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
         'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
         'parking', 'prefarea', 'furnishingstatus', 'price']]
```

```python
for i in df.columns.tolist():
    if df[i].dtype == 'object':
        df[i] = df[i].apply(lambda x: 0 if x in ['no','unfurnished'] else(1 if x in ['yes','furnished'] else 0.5))
```

```python
from src import LoadData
```

```python
ld = LoadData('housing-data.csv').load_csv()
```

```python
X_train, X_val, y_train, y_val = ld.train_val_split(randomize=False)
```

```python
X_train, X_val = ld.standardize(X_train,X_val)
```

```python
y_mean = y_train.mean()
y_std = y_train.std()

y_train_scaled = (y_train - y_mean) / y_std
y_val_scaled   = (y_val - y_mean) / y_std
```

```python
X_train, X_val = ld.add_bias_term(X_train,X_val)
```

```python
np.save('X_train.npy',X_train)
np.save('y_train.npy',y_train_scaled)
np.save('X_val.npy',X_val)
np.save('y_val.npy',y_val_scaled)
```
