**Importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
from predictor import Predictor
```

```python
# Generate synthetic regression data
X,y = datasets.make_regression(n_samples=100, # number of samples
    n_features=1, # one feature (independent variable)
    noise=20, # add noise for realism
    random_state=4) # reproducibility
# Spliting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y,
    test_size=0.2, # 20% for testing
    random_state=1234 # reproducibility
    )
```

**Custom Scatter Plot**

```python
# Create figure with a custom size and outer background color
fig=plt.figure(figsize=(10.24,7.68))
fig.set_facecolor("#b1b1b1")# Setting the background color
# Setting up a white horizontal grid with specific style
plt.grid(axis='y', color='white', linestyle='-', linewidth=1, alpha=0.3)
plt.gca().set_facecolor("#2b2b2b")
plt.scatter(X[:,0],y,color="r",marker="*",s=30)
# Optimize the layout and display the plot
plt.tight_layout()
plt.show()
```

**Linear Regression Prediction and Manual MSE Calculation**

```python
linreg = Predictor()  # Custom regression model
linreg.model(X_train,y_train) # Trains the model
# Predicting outcomes for test data
predictions = linreg.predict(X_test)
# Define a function
def mse(y_test, predictions):
    return np.mean((y_test-predictions)**2)
# Compute and print MSE
mse = mse(y_test,predictions)
print(mse)
```

**Visualizing Linear Regression Results**

```python
# Predict values across all X for regression line
y_pred_line = linreg.predict(X)
# Creating a figure and set the outer background color
fig=plt.figure(figsize=(10.24,7.68))
fig.set_facecolor("#7e7e7e")
# Plot training data
m1 = plt.scatter(X_train,y_train,color='magenta',s=12,marker='x',label='Training data')
m2 = plt.scatter(X_test,y_test,color='yellow',s=15,marker='*',label='Test data')
# Setting up a grid on y-axis with white lines
plt.grid(axis='y', color='white', linestyle='-', linewidth=1, alpha=0.3)
plt.gca().set_facecolor("#2b2b2b")
# Plot regression prediction line
plt.plot(X,y_pred_line,color='cyan',linewidth=1,ls=':',label='Prediction')
# Optimize layout, add legend, and display
plt.tight_layout()
plt.legend()
plt.show()
```
