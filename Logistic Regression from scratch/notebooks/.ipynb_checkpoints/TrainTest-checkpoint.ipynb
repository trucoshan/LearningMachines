# Training and Testing

```python
import sys
sys.path.append('../')
from src import LogisticRegression

import numpy as np

import os
os.chdir('c:\\Users\\shant\\Desktop\\Logistic Regression from scratch\\train')
X_train = np.load('X_train.npy')
y_train = np.load('y_train.npy')
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
os.chdir('c:\\Users\\shant\\Desktop\\Logistic Regression from scratch\\test')
X_val = np.load('X_val.npy')
y_val = np.load('y_val.npy')
```

```python
y_pred = model.predict(X_val)
```

```python
errs = model.evaluate(y_pred,y_val)
```
